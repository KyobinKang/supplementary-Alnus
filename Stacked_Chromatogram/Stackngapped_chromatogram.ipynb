## Draw stacked chromatograms with preset interval between samples

**Author**: Kyo Bin Kang (kbinkang@gmail.com) <br>

```r
library("ggplot2")
```

```r
rawdata <- read.csv("Raw_chromatogram.csv")
```

Give identifier to each unique species / plant part

```r
species <- 1:length(unique(rawdata$Species))
names(species) <- unique(rawdata$Species) 
parts <- 1:length(unique(rawdata$Parts))
names(parts) <- unique(rawdata$Parts)
```

Set the interval between samples

```r
gap_species <- 25000
gap_parts <- 100000
```

Place the interval between chromatograms and save it as 'rawdata2'

```r
rawdata2 <- rawdata
rawdata2$BPI <- rawdata2$BPI + gap_species*species[rawdata$Species] + gap_parts*parts[rawdata$Parts]
```

Draw and check the chromatograms

```r
ggplot(rawdata2, aes(x = RT, y = BPI, color = Species, linetype = Parts)) +
        geom_line() +
        labs(x = 'Retention time (min)',
             y = '') +
        theme_bw() +
        theme(panel.grid.major.x = element_blank(), 
            panel.grid.minor.x = element_blank(),
            panel.grid.major.y = element_blank(),
            panel.grid.minor.y = element_blank()) +
        xlim(0,25)
```

```r
#save plot using GGplot2
myplot <- ggplot(rawdata2, aes(x = RT, y = BPI, color = Species, linetype = Parts)) +
        geom_line() +
        labs(x = 'Retention time (min)',
             y = 'Intensity') +
        theme_bw() +
        theme(panel.grid.major.x = element_blank(), 
            panel.grid.minor.x = element_blank(),
            panel.grid.major.y = element_blank(),
            panel.grid.minor.y = element_blank()) +
        theme(axis.text.x = element_text(colour="grey20", size=20, angle=90, hjust=.5, vjust=.5),
            axis.text.y = element_text(colour="grey20", size=20),
            text=element_text(size=20)) +
        xlim(0,25)
ggsave("Alnus_chromatogram_plantparts.png", myplot, width=20, height=10)
```
